```python
import neurodesk as nd
import os
```

```python
# Lookup a container from CVMFS by name.
# This lookup will fail if the container does not exist in CVMFS.
niimath = nd.container("niimath", debug=True)
```

```python
# Run the exported command "niimath" in the "niimath" container.
out = niimath.niimath()

out.split("\n")[0]
```

```python
print(niimath.bash("-c", "ls -la"))
```

```python
shared = nd.share_dir(os.getcwd())
```

```python
print(niimath.shell("cat", shared / "hello.txt"))
```
